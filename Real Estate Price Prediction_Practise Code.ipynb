```python
X=data.drop("Y house price of unit area",axis=1)
Y=data["Y house price of unit area"]
```

```python
data_corr=data.corr()
plt.subplots(figsize=(10,8))
sns.heatmap(data_corr,vmin=0.4,
            vmax=0.8,
            annot=True,
            square=True,
            fmt='.1f',
            cmap = 'Purples')
```

```python
import statsmodels.api as sm
res=sm.OLS(Y,X)
mod=res.fit()
print(mod.summary())
```

```python
# Since All variable are statistically significant
```

```python
# 70-30 split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
model=linreg.fit(X_train,Y_train)
```

```python
print('R-squared value for training dataset is %.3f' %model.score(X_train,Y_train))
print('R-squared value for testing dataset is %.3f' %model.score(X_test,Y_test))
print('The estimated model intercept is %.3f' %model.intercept_)
```

```python
model.coef_
```

```python
# Prediction on training and testing datasets
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
pred_all = model.predict(X)

#Calculating Mean Squared Error (MSE)
print('MSE of training dataset: %.3f' %np.mean((Y_train - pred_train)**2))
print('MSE of testing dataset: %.3f' %np.mean((Y_test-pred_test)**2))
print('MSE of overall dataset: %.3f' %np.mean((Y-pred_all)**2))
```

```python
# PLoting scatter plot for residuals in train and test to see if there is any pattern
train = plt.scatter(pred_train,(Y_train-pred_train),color='blue',alpha=0.5,label='TRAINING')
test = plt.scatter(pred_test,(Y_test-pred_test),color='red',alpha=0.5,label="TESTING")
plt.hlines(y=0,xmin=-10,xmax=100000)
plt.title('RESIDUAL PLOT')
plt.xlabel('Predicted points')
plt.ylabel('Residual points')
plt.legend(loc=3)
```

```python
 # There is no pattern in residual so the model is correctly captured
```

```python
# Creating Predected_target Ranking comparision of target and predicted variables
data['Predicted_Price'] = pd.DataFrame(pred_all).round(decimals=2)
data['Target_rank'] = pd.qcut(data['Y house price of unit area'].rank(method='first').values,5,duplicates='drop').codes+1
data['Predicted_rank'] = pd.qcut(data['Predicted_Price'].rank(method='first').values,5,duplicates='drop').codes+1
data.head()
```

```python
# Analysing the prediction by pivot table
pivot_table = pd.pivot_table(data,values='Y house price of unit area', index='Predicted_rank', columns='Target_rank',aggfunc='count')
pivot_table
```
